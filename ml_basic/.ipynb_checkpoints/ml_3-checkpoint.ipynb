### 분류 모델의 성능 평가
- 정확도 : 실 데이터와 예측 데이터가 얼마나 같은지 판단하는 지표
- 문제점? 이진분류의 경우 모델의 성능을 왜곡할 수 있다.
- 왜 : 데이터의 불균형 -> 한쪽으로 치우친 데이터를 예측하는 경우 문제가 된다.
- 해결책 : F1 Score(Precision, Recall)

```python
print('TP(target = 1, predict = 1)')
print('TN(target = 0, predict = 0)')
print('FN(target = 1, predict = 0 -> Type 1 error)')
print('FP(target = 0, predict = 1 -> Type 2 error)')

target = [1,0,0,1,1,1,0,1,1,1]
prediction = [0,1,1,1,1,0,1,0,1,0]

tp = tn = fn = fp = 0

for idx in range(len(target)):
    # tp
    if target[idx] == 1 and prediction[idx] == 1:
        tp += 1
    elif target[idx] == 1 and prediction[idx] == 0:
        fn += 1
    elif target[idx] == 0 and prediction[idx] == 1:
        fp += 1
    else:
        tn += 1
        
print('tp : \n', tp)
print('tn : \n', tn)
print('fn : \n', fn)
print('fp : \n', fp)
```

```python
print('accuracy - ', accuracy_score(target, prediction))
print('recall - ', recall_score(target, prediction))
print('precision - ', precision_score(target, prediction))
print('f1 score - ', f1_score(target, prediction))
print('confuse matrix - ', confusion_matrix(target, prediction))
```

- 정밀도(Precision : tp / (fp + tp)
- 상대적으로 정밀도가 더 중요한 지표인 경우의 모델? - 스팸메일


- 재현율(Recall) : TP / (FN + TP)
- 상대적으로 재현율이 더 중요한 지표인 경우의 모델? - 의학(암진단 등), 금융 사기 등

```python
titanic_frm = pd.read_csv('./lecture_file/titanic_train.csv')
titanic_frm.describe()
```

```python
titanic_frm['Survived'].value_counts()
```

```python
print('2. target, feature로 데이터 분리')

titanic_target = titanic_frm.Survived
titanic_feature = titanic_frm.drop(labels='Survived', axis=1)
# titanic_target.values
```

```python
print('3. 전처리 요구사항 - ')
print('불필요한 피처제거 - PassengerId, Name, Ticket')
print('결측값 처리       - AGE는 평균, Cabin는 N, Embraked는 N')
print('Cabin의 경우 앞 문자 하나에 대해서 레이블 인코딩만 진행!!')
print('레이블 인코딩     - Sex, Cabin, Embarked')

titanic_feature = titanic_feature.drop(columns=['PassengerId', 'Name', 'Ticket'],axis = 1)
titanic_feature.isnull().sum()
titanic_feature
```

```python
titanic_feature['Age'].fillna(titanic_feature.Age.mean(), inplace=True)
titanic_feature['Cabin'].fillna('N', inplace=True)
titanic_feature['Embarked'].fillna('N', inplace=True)
```

```python
def label_encode(data):
    data['Cabin'] = data['Cabin'].str[:1]
    features = ['Sex', 'Cabin', 'Embarked']
    
    for i in features:
        le = LabelEncoder()
        data[i] = le.fit_transform(data[i])

    return data
```

```python
feature_subset = label_encode(titanic_feature)
feature_subset
```

```python
print('학습과 테스트 분리 - ')

train_x, test_x, train_y, test_y = train_test_split(feature_subset, titanic_target,
                                                   test_size= 0.2,
                                                   random_state=200)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
lr_model = LogisticRegression()
lr_model.fit(train_x, train_y)
lr_y_pred = lr_model.predict(test_x)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(train_x, train_y)
dt_y_pred = dt_model.predict(test_x)
```

```python
def metrics_eval(target, prediction): 
    print('accuracy  - ', accuracy_score(target, prediction))
    print('recall    - ', recall_score(target, prediction))
    print('precision - ', precision_score(target, prediction))
    print('f1 score  - ', f1_score(target, prediction))
    print('confuse matrix - ', confusion_matrix(target, prediction))
```

```python
print('LogisticRegression - \n')
metrics_eval(test_y, lr_y_pred)
```

```python
print('DecisionTreeClassifier - \n')
metrics_eval(test_y, dt_y_pred)
```

```python
print('교차검증 - ')
fold = KFold(n_splits=20)
scoring = {
    'accuracy' : make_scorer(accuracy_score),   # make_scorer는 함수를 인자로 받음
    'precision': make_scorer(precision_score),
    'recall'   : make_scorer(recall_score),
    'f1'       : make_scorer(f1_score)
}

lr_result = cross_validate(lr_model, train_x, train_y,
                       cv = fold,
                       scoring=scoring)

dt_result = cross_validate(dt_model, train_x, train_y,
                       cv = fold,
                       scoring=scoring)
lr_result.keys()
```

```python
lr_result['test_accuracy'].mean()
```

```python
dt_result['test_accuracy'].mean()
```

```python
print('유방암 관련 데이터 - 정확도, 재현율(실제 P를 N으로 예측하면 안되기 때문에)')
print('재현율은 실제 양성을 양성으로 예측한 비율이 높아야 성능이 좋은 모델')
load_breast_cancer().keys()
```

### 실습

```python
print('1. 데이터프레임 만들기')
print('2. target에 대한 균형 여부 확인')
print('3. 데이터 세트 분리')
print('4. RandomForestClassifier')
print('5. 평가지표 확인')
```

```python
print('1. 데이터프레임 만들기')
cancer = load_breast_cancer()
df_data = pd.DataFrame(cancer.data)
df_target = pd.DataFrame(cancer.target)
```

```python
print('2. target에 대한 균형 여부 확인')
df_target.value_counts()
```

```python
print('3. 데이터 세트 분리')
train_x, test_x, train_y, test_y = train_test_split(df_data, df_target,
                                                   test_size=0.2)
```

```python
print('4. RandomForestClassifier')
rfc = RandomForestClassifier(random_state=200)

rfc.fit(train_x, train_y)
rfc_pred = rfc.predict(test_x)
```

```python
print('5. 평가지표 확인')
print(metrics_eval(test_y, rfc_pred))
```

```python
print('6. 교차검증 확인')

fold = KFold(n_splits=20)
scoring = {
    'accuracy' : make_scorer(accuracy_score),   # make_scorer는 함수를 인자로 받음
    'precision': make_scorer(precision_score),
    'recall'   : make_scorer(recall_score),
    'f1'       : make_scorer(f1_score)
}

rfc_result = cross_validate(rfc, train_x, train_y,
                       cv = fold,
                       scoring=scoring)
```

```python
rfc_result['test_recall'].mean()
```

```python
print('재현율을 높이기 위한 방법으로 - GridSearchCV를 이용한 파라미터 튜닝!!')
print('n_estimators - tree 갯수를 의미')
print('max_features - 최대 선택할 피처의 수를 의미')
print('max_depth    - 최대 선택할 트리의 깊이를 의미')
```

```python
param = {
    'n_estimators' : [50, 100, 150, 200],
    'max_features' : [6, 8, 15, 20],
    'max_depth'    : [4, 6, 8]
}

grid_search_model = GridSearchCV(rfc,
                                 param_grid=param,
                                 cv = 20,
                                 refit=True,
                                 scoring='recall')

grid_search_model.fit(train_x, train_y)
```

```python
print(grid_search_model.best_score_, grid_search_model.best_params_)
```

```python
rfc_best_model = grid_search_model.best_estimator_
```

```python
rfc_best_pred = rfc_best_model.predict(test_x)
```

```python
metrics_eval(test_y, rfc_best_pred)
```

### 실습  위에 내용을 문턱을 걸고 학습 및 검증해보기

```python
titanic_frm = pd.read_csv('./lecture_file/titanic_train.csv')
titanic_frm.describe()

print('2. target, feature로 데이터 분리')

titanic_target = titanic_frm.Survived
titanic_feature = titanic_frm.drop(labels='Survived', axis=1)
# titanic_target.values

print('3. 전처리 요구사항 - ')
print('불필요한 피처제거 - PassengerId, Name, Ticket')
print('결측값 처리       - AGE는 평균, Cabin는 N, Embraked는 N')
print('Cabin의 경우 앞 문자 하나에 대해서 레이블 인코딩만 진행!!')
print('레이블 인코딩     - Sex, Cabin, Embarked')

titanic_feature = titanic_feature.drop(columns=['PassengerId', 'Name', 'Ticket'],axis = 1)
titanic_feature.isnull().sum()
titanic_feature

titanic_feature['Age'].fillna(titanic_feature.Age.mean(), inplace=True)
titanic_feature['Cabin'].fillna('N', inplace=True)
titanic_feature['Embarked'].fillna('N', inplace=True)

def label_encode(data):
    data['Cabin'] = data['Cabin'].str[:1]
    features = ['Sex', 'Cabin', 'Embarked']
    
    for i in features:
        le = LabelEncoder()
        data[i] = le.fit_transform(data[i])

    return data

feature_subset = label_encode(titanic_feature)
feature_subset

print('학습과 테스트 분리 - ')
train_x, test_x, train_y, test_y = train_test_split(feature_subset, titanic_target,
                                                   test_size= 0.2,
                                                   random_state=200)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(train_x, train_y)
y_pred = logistic_model.predict(test_x)

print('확률예측 값 - predict_proba()')
predict_proba_result = logistic_model.predict_proba(test_x)

print('type - ', type(predict_proba_result))
print('shape - ', predict_proba_result.shape)
```

```python
pred_prob_concat = np.concatenate([predict_proba_result , y_pred.reshape(-1, 1)], axis = 1)
pred_prob_concat[:10]
```

```python
user_th = 0.3

# positive array
predict_proba_result[:,1]

predict_proba_positive = predict_proba_result[:,1].reshape(-1,1)

# 임계값 변경하여 적용
user_pred = Binarizer(threshold = user_th).fit_transform(predict_proba_positive)
```

```python
print('default - 0.5 \n')
metrics_eval(test_y, y_pred)

print('\n user th - 0.3 \n')
metrics_eval(test_y, user_pred)
```

### trade off 시각화
- precision_recall_curve(실제값, 예측 확률 값) : 임계값 변화에 따른 평가지표를 반환
- 반환값 : 정밀도, 재현율, 임계값

```python
predict_proba_positive = predict_proba_result[:, 1]
precision, recall, th = precision_recall_curve(test_y, predict_proba_positive)

print('precision - ', type(precision))
print('recall - ', type(recall))
print('th - ', type(th))
print('precision - ', precision.shape)
print('recall - ', recall.shape)
print('th - ', th.shape)
```

```python
plt.figure(figsize= (15, 5))

plt.plot(th, precision[:165], linestyle = '--', label = 'precision')
plt.plot(th, recall[:165], linestyle = '-', label = 'recall')

plt.grid()
plt.legend()
plt.xlabel('threshold ratio')
plt.ylabel('precision and recall value')
plt.show()
plt.close()
```
